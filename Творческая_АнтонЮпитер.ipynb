# NETWORKX
Networkx - это проект, который помогает пользователю в работе с графами. Графы - это форма визуализации, которая позволяет анализировать
и показывать отношения между обьектами.


Графы, т.е. Networkx используются в разных областях. Например, в  фундаментальных научных исследованиях и в прикладных задачах. Простыми словами - графы используются как и в науке, так и в быту.





```python
import networkx as nx
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.nodes()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 4)
G.nodes()
nx.draw(G)
```

## ВОТ ПОЧЕМУ ВЫ ПОЛЮБИТЕ NETWORKX

NetworkX имеет встроенные алгоритмы, позволяющие выполнять типовые операции с графами, такие как: нахождение кратчайшего пути, кластеризация, поиск в высоту и ширину.

## В NetworkX вы сможете сделать графы разных типов:

### Простые неориентированные 
B таких графах вершины соединяются одиночными ребрами, неимеющими вектора направленности (типичный пример — двухполосная дорожная система);


### Ориентированные
B графах этого типа связи между узлами имеют направление (примером такой структуры является река с притоками);

### Мультиграфы
B них парные вершины соединены более, чем одним ребром, также допускается наличие петель (ребер, замыкающихся на один и тот же узел);


### Ориентированные мультиграфы
B таких графах соседние узлы соединяются несколькими направленными ребрами.
